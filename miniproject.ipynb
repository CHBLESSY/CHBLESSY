```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars=pd.read_csv("https://raw.githubusercontent.com/arora123/Data/master/Automobile_data.csv")
cars.head()
```

```python
print(cars.isnull().sum())
cars_cleaned = cars.dropna()
print(f"Data after dropping missing values: {cars_cleaned.shape}")
```

```python
min_price = cars_cleaned['price'].min()
max_price = cars_cleaned['price'].max()

cars_min_price = cars_cleaned[cars_cleaned['price'] == min_price]
cars_max_price = cars_cleaned[cars_cleaned['price'] == max_price]

print("Cars with minimum price:")
print(cars_min_price[['company', 'body-style', 'price']])

print("Cars with maximum price:")
print(cars_max_price[['company', 'body-style', 'price']])
```

```python
body_style_count = cars_cleaned['body-style'].value_counts()
print("Count of cars by body style:")
print(body_style_count)
```

```python
avg_price_body_style = cars_cleaned.groupby('body-style')['price'].mean().sort_values(ascending=False)
print("Average price by body style:")
print(avg_price_body_style)
```

```python
avg_price_company = cars_cleaned.groupby('company')['price'].mean().sort_values(ascending=False)
print("Average price by company:")
print(avg_price_company)
```

```python
cylinders_count = cars_cleaned['num-of-cylinders'].value_counts()
print("Count of cars by number of cylinders:")
print(cylinders_count)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='company', y='price', data=cars_cleaned)
plt.xticks(rotation=90)
plt.title('Company vs Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='company', hue='num-of-cylinders', data=cars_cleaned)
plt.xticks(rotation=90)
plt.title('Company vs Number of Cylinders')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='body-style', y='price', data=cars_cleaned)
plt.xticks(rotation=90)
plt.title('Body Style vs Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='body-style', hue='num-of-cylinders', data=cars_cleaned)
plt.xticks(rotation=90)
plt.title('Body Style vs Number of Cylinders')
plt.show()
```
